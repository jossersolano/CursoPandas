# Pandas II: joins y fechas

```python
import pandas as pd
```

```python
# carga de datos
df1 = pd.read_csv('../data/raw/dataset_2a.csv')
df2 = pd.read_csv('../data/raw/dataset_2b.csv')
df3 = pd.read_csv('../data/raw/dataset_2c.csv')
```

```python
df1
```

```python
df2
```

```python
df3
```

```python
# join
df_con_usuarios = df1.merge(df2, on='user', how='left')
df_con_usuarios
```

```python
df_completo = df_con_usuarios.merge(df3, on='ciudad', how='left')
df_completo
```

```python
# trato de fechas
df_completo['fecha'] = pd.to_datetime(df_completo['fecha_publicado'], format='%d/%m/%Y')
df_completo
```

```python
# revisar esto https://docs.python.org/3/library/datetime.html
df_completo['dia_semana'] = df_completo['fecha'].apply(lambda x: x.strftime('%A'))
df_completo
```

```python
df_completo['mes'] = df_completo['fecha_publicado'].apply(lambda x: x[3:])
df_completo
```

```python
# Agrupacion del mes
df_completo.groupby('mes').agg({
    'id': 'count'
})
```

```python
# Agrupacion m√∫ltiple
agrupado = df_completo.groupby(['mes', 'ciudad']).agg({
    'id': 'count'
})
agrupado
```

```python
# pilas con los indices, no columnas
# Debemos reseterar el indice para que se convierta en una columnas
desagrupado = agrupado.reset_index(1) # resetear ciudad, nivel 1
desagrupado
```

```python
desagrupado['ciudad']
```

```python
# ahpra a pivotear
pivote = desagrupado.pivot(columns='ciudad')
pivote
```

```python
# Ahora a desnivelar las columnas
pivote = pivote.droplevel(0, axis='columns')
pivote
```

```python
pivote[['GUAYAQUIL']]
```

```python
pivote = pivote.fillna(0)
pivote
```

```python
# funci√≥n de transponer
pivote.transpose()
```
