# Pandas II: joins y fechas

```python
import pandas as pd
```

```python
# carga de datos
df1 = pd.read_csv('../data/raw/dataset_2a.csv')
df2 = pd.read_csv('../data/raw/dataset_2b.csv')
df3 = pd.read_csv('../data/raw/dataset_2c.csv')
```

```python
df1
```

```python
df2
```

```python
df3
```

```python
# join
df_con_usuarios = df1.merge(df2, on='user', how='left')
df_con_usuarios
```

```python
df_completo = df_con_usuarios.merge(df3, on='ciudad', how='left')
df_completo
```

```python
# trato de fechas
df_completo['fecha'] = pd.to_datetime(df_completo['fecha_publicado'], format='%d/%m/%Y')
df_completo
```

```python
# revisar esto https://docs.python.org/3/library/datetime.html
df_completo['dia_semana'] = df_completo['fecha'].apply(lambda x: x.strftime('%A'))
df_completo
```

```python
df_completo['mes'] = df_completo['fecha_publicado'].apply(lambda x: x[3:])
df_completo
```
