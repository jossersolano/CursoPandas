```python
import pandas as pd

# df = pd.read_csv("../data/raw/dataset.csv")

df = pd.read_csv("../data/raw/dataset.csv", index_col="id")

# df # imprime el dataframe completo si es posible

df.head() # imprime los 5 primeros y dentro del () puedes redefinir las cantidades de filas

# df.tail(10) # imprime los 5 ultimos y en () puedes redefinir la cantidad

# df.describe() # aqui te imprime lo que es la cantidad, desviacion standar?, min, max, percentiles
```

```python
# df_filtrado = df.dropna() # aqui elimina las filas que tienen datos vacios o NaN
# df_filtrado = df.fillna(0) # aqui trasnformas esos NaN en 0

df_filtrado = df.fillna({"retweets": 0, "mentions": -1}) # aqui especificas/redefines los valores en cada columna en vez de cero

df_filtrado.head()
```

```python
# df_filtrado["favorites"]

df_filtrado[["favorites", "full_text"]]
```

```python
# df_filtrado.iloc[0] # iloc es el filtro por indice

# df_filtrado.iloc[0:3] # tambien puede ser un rango

# df_filtrado.iloc[[0, 2, 1]] # incluso los indices seleccionados

# df_filtrado.loc[[183744, 183746]] # si no se quiere tambien puede ser por el id que se definió pero hay que quitar el i de iloc para quedar loc

df_filtrado.loc[[183741, 183742], ["favorites", "full_text"]] # tambien se filtra por filas (id) y columnas
```

```python
# y ahora filtrado por condiciones

# df_filtrado[df_filtrado["favorites"] > 400]

# df_filtrado[(df_filtrado["favorites"] > 400) & (df_filtrado["mentions"] > 20)] # filtra por varias condiciones, no olvidar los parentesis

df_filtrado[df_filtrado["full_text"].str.contains("Programming")]
```

```python
# creando nueva columna o trasnformarla
import random
def calcular_ganacias(retweets):
    ganancia = retweets * random.randint(3, 5)
    return ganancia

df_filtrado["ganancias"] = df_filtrado["retweets"].apply(calcular_ganacias)

df_filtrado.head(10)
```

```python
# creando una columna calculando 2 columnas
def popularidad(fila):
    resultado = fila["followees"]/fila["followers"]
    return resultado

df_filtrado["popularidad"] = df_filtrado.apply(popularidad, axis=1)

df_filtrado.head()
```

```python
# agrupando datos
# df_filtrado.groupby("country").mean() #esto me genera error porque ahora debo excluir la columna de texto

# aqui me ayudo la IA
import numpy as np # importo numpy

df_numerico = df_filtrado.select_dtypes(include=[np.number]) # selecciono solo las columnas que tienen números
df_numerico['country'] = df_filtrado['country'] # ahora agrego la columna de texto del df original
df_numerico.groupby("country").mean() # ahora si puedo agregar. Creo que mas facil era quitar la columna de full_text
```

```python
# ahora agrupo por varios conceptos
grouped = df_filtrado.groupby("country").agg({
    "followers": 'sum',
    "mentions": 'mean',
    "retweets": 'max'
})

grouped[grouped["followers"] > 5000]
```

```python
# y para guarda los datos
grouped.to_csv('../data/processed/salida.csv')
```
